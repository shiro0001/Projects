```python
# Importing required library 
import requests
import pandas as pd
import json

# ---Getting Movie ID---
def get_movie_ID(movieName,key):
    # API endpoint 
    url = "https://api.themoviedb.org/3/search/movie"
    
    # Parameters for API request
    params = {
        "api_key": key,
        "query": movieName
    }
    
    # Send API request and get response
    response = requests.get(url, params=params)
    data = response.json()
    
    # Get the first movie result and return its ID
    if data["total_results"] > 0: # if there is value to this movie name input
        movie_id = data["results"][0]["id"]
        return movie_id
    else:
        return(f"No results found for {movieName}.\n***There might have been error to the way the movie name is written***")
    
    
    
#---Getting Movie Reccomendation---
def movie_Reccomendation(movie_id,key):

    # End point for reccomendation API
    recc_url = f"https://api.themoviedb.org/3/movie/{movie_id}/recommendations?api_key={key}&language=en-US&page=1"
    
    # Send API request and get response
    r = requests.get(recc_url)
    data = r.json()
    
    for movie in data["results"]:
        print(movie["title"])
    
    
    
# The main method
def main():
    API_Key ="[your api key]"
    cont=True
    while cont== True:
        
        #I assume that the movie name is always type correctly 
        user_Input= input("Whats your favorite movie ")

        #Getting movie ID
        movieID = get_movie_ID(user_Input,API_Key)
        
        # Getting movie reccomendation
        print("Some movie you might like")
        movie_Recc = movie_Reccomendation(movieID,API_Key)
        
        
        go = input('enter "g" to continue, type anything else to stop ').upper()
        if go=="G":
            count=True
        else:
            cont=False

main()
```
