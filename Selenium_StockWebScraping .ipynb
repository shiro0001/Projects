```python
#import relevant library (selenium)
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait as wait
from selenium.webdriver.support import expected_conditions as EC

# importing library for web scraping
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import json
import time



#---scraping tech stock name for selenium---

# starting scraping process
url =  "https://en.wikipedia.org/wiki/Big_Tech"
response = requests.get(url)
soup = bs(response.text, "html.parser")

#finding all table on the website to scrape with the class listed below
table = soup.find_all("table", {"class": "wikitable sortable mw-collapsible floatright"})

# Read the html table and put the dataframe in a list
df = pd.read_html(str(table))


# Getting the 1st DF named 'Big Five tech companies'
tb1 =df[0]
# Getting the 2nd DF named 'Smaller U.S. Big Tech companies'
tb2 =df[1]
DF1 = pd.concat([tb1,tb2])

company_Name = DF1["Company"].tolist()
print(company_Name)

```

```python
#---selenium part---

#Create a Chrome web driver
driver = webdriver.Chrome()
actions = ActionChains(driver)

#Navigate to website
driver.get("https://finance.yahoo.com/")

# search for the full popup and closing it
ads_search = driver.find_element(By.XPATH, '//*[@id="overlay-module"]')
close_button = driver.find_element(By.XPATH, '//*[@id="myLightboxContainer"]/button')
close_button.click()

#Insantiating list
Stock_Name = []
previousClose = []
openingPrice = []
dailyPriceRange = []
week52PriceRange = []
volume = []
marketCap = []
beta = []
dividend = []
nextDividendDate = []



for stocks in company_Name:
    #Locate the search bar and send the keyword to it
    search = driver.find_element(By.ID, 'yfin-usr-qry')
    search.send_keys(stocks)

    #Find the search button on the html and click search
    search_button = driver.find_element(By.ID, 'header-desktop-search-button')
    search_button.click()
    time.sleep(5)


    #Finding all value in the table
    table_items = wait(driver, 10).until(EC.presence_of_all_elements_located((By.XPATH, '//*[@id="quote-summary"]')))

    for item in table_items:
    #Getting previous close
        PC = item.find_element(By.XPATH, '//*[@id="quote-summary"]/div[1]/table/tbody/tr[1]/td[2]')
        value1 = PC.text
        previousClose.append(float(value1))

    #Getting Open
        OP = item.find_element(By.XPATH,'//*[@id="quote-summary"]/div[1]/table/tbody/tr[2]/td[2]')
        value2 = OP.text
        openingPrice.append(float(value2))

    #Getting days range
        DR = item.find_element(By.XPATH,'//*[@id="quote-summary"]/div[1]/table/tbody/tr[5]/td[2]')
        value3 = DR.text
        dailyPriceRange.append(str(value3))

    #Getting 52 week range
        WR52 = item.find_element(By.XPATH,'//*[@id="quote-summary"]/div[1]/table/tbody/tr[6]/td[2]')
        value4 = WR52.text
        week52PriceRange.append(str(value4))

    #Getting volume
        V = item.find_element(By.XPATH,'//*[@id="quote-summary"]/div[1]/table/tbody/tr[7]/td[2]/fin-streamer')
        value5 = V.text
        volume.append(str(value5))

    #Getting market cap
        MC = item.find_element(By.XPATH,'//*[@id="quote-summary"]/div[2]/table/tbody/tr[1]/td[2]')
        value6 = MC.text
        marketCap.append(str(value6))

    #Getting Beta(closer to 1 are generally good, less risk)
        B = item.find_element(By.XPATH,'//*[@id="quote-summary"]/div[2]/table/tbody/tr[2]/td[2]')
        value7 = B.text
        beta.append(float(value7))

    #Getting dividend yield 
        DDD = item.find_element(By.XPATH,'//*[@id="quote-summary"]/div[2]/table/tbody/tr[6]/td[2]')
        value8 = DDD.text
        dividend.append(str(value8))

    #Getting next ex dividend date
        NDD = item.find_element(By.XPATH,'//*[@id="quote-summary"]/div[2]/table/tbody/tr[7]/td[2]')
        value9 = NDD.text
        nextDividendDate.append(str(value9))
        
    # Adding stock name to list
    Stock_Name.append(stocks)
        
    # printing the list result
    print(f'pc = {previousClose}')
    print(f'op = {openingPrice}')
    print(f'daily price range = {dailyPriceRange}')
    print(f'52 week = {week52PriceRange}')
    print(f'Volume = {volume}')
    print(f'market cap = {marketCap}')
    print(f'Beta = {beta}')
    print(f'Dividend = {dividend}')
    print(f'next Dividend Date = {nextDividendDate}')


    # clearing search box 
    driver.find_element(By.ID, 'yfin-usr-qry').clear()

    # wait 5 sec
    time.sleep(5)

    # waiting 5 seconds for the page to load
    driver.implicitly_wait(5)

driver.quit()
```

```python
# making the data frame
stockinfo_DF = pd.DataFrame({
    
    "Stocks (Ticker)": Stock_Name,
    "Previous Close": previousClose,
    "Opening Price": openingPrice,
    "Daily Price Range" : dailyPriceRange,
    "Week 52 Price Range" : week52PriceRange,  
    "Volume"  : volume,
    "Market Cap" : marketCap,
    "Beta": beta,
    "Dividend this month" : dividend,
    "Next Dividend Date" : nextDividendDate 
    
})


stockinfo_DF
```
